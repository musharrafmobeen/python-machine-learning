```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('tensorflowData/fake_reg.csv')
```

```python
df
```

```python
sns.pairplot(df)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['feature1','feature2']].values
```

```python
y = df['price'].values
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_test
```

```python
# model = Sequential([Dense(4,activation = 'relu'),Dense(2,activation= 'relu'),Dense(1)])
```

```python
model = Sequential()
```

```python
model.add(Dense(4,activation='relu'))
model.add(Dense(4,activation='relu'))
model.add(Dense(4,activation='relu'))
model.add(Dense(1))
          
model.compile(optimizer='rmsprop',loss='mse')          
```

```python
model.fit(x=X_train,y=y_train,epochs=250)
```

```python
model.history.history
```

```python
loss_dataframe = pd.DataFrame(model.history.history)
```

```python
loss_dataframe
```

```python
loss_dataframe.plot()
```

```python
model.evaluate(x=X_test,y=y_test,verbose = 0)
```

```python
model.evaluate(x=X_train,y=y_train,verbose=0)
```

```python
test_predictions = model.predict(X_test)
```

```python
test_predictions
```

```python
from sklearn import metrics
```

```python
test_predictions = pd.Series(test_predictions.reshape(300,))
pred_df = pd.DataFrame(y_test,columns=['Test True Y'])
```

```python
pred_df = pd.concat([pred_df,test_predictions],axis=1)
```

```python
pred_df
```

```python
pred_df.columns = ['Test True Y','Model Predictions']
```

```python
pred_df
```

```python
sns.scatterplot(x='Test True Y',y='Model Predictions',data=pred_df)
```

```python
metrics.mean_absolute_error(pred_df['Test True Y'],pred_df['Model Predictions'])
```

```python
metrics.mean_squared_error(pred_df['Test True Y'],pred_df['Model Predictions'])
```
